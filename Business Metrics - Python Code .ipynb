```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = pd.read_csv('C:/Users/amith/Downloads/^IXIC.csv')

# Calculate the correlation between GOOGL and IXIC
correlation = data['Close(GOOGLE)'].corr(data['Close(IXIC)'])

# Create a scatter plot to show the correlation
plt.figure(figsize=(10, 6))
plt.scatter(data['Close(GOOGLE)'], data['Close(IXIC)'], color='b')
plt.title('Scatter Plot of GOOGL vs IXIC Closing Prices')
plt.xlabel('GOOGL Closing Price')
plt.ylabel('IXIC Closing Price')
plt.grid(True)
plt.show()

print(f"The correlation between GOOGL and IXIC is: {correlation:.2f}")
```

The analysis presented on the scatter plot suggested a positive correlation between Google's stock price (GOOGL) and the closing price of the Nasdaq Index (IXIC). A correlation coefficient of 0.83, displayed within the image, indicated a strong positive trend. This meant that, as the closing price of the Nasdaq Index increased, Google's stock price also tended to rise, and vice versa

```python
# Calculate daily percentage change in closing prices
data['Daily Change(GOOGLE)'] = data['Close(GOOGLE)'].pct_change() * 100
data['Daily Change(IXIC)'] = data['Close(IXIC)'].pct_change() * 100

# Plotting the closing prices
plt.figure(figsize=(14, 7))

# GOOGL closing prices
plt.subplot(2, 1, 1)
plt.plot(data['Days(GOOGLE)'], data['Close(GOOGLE)'], label='GOOGL', color='b')
plt.title('GOOGL vs IXIC Stock Performance')
plt.xlabel('Date')
plt.ylabel('Closing Price (GOOGL)')
plt.legend()

# IXIC closing prices
plt.subplot(2, 1, 2)
plt.plot(data['Days(IXIC)'], data['Close(IXIC)'], label='IXIC', color='r')
plt.xlabel('Date')
plt.ylabel('Closing Price (IXIC)')
plt.legend()

plt.tight_layout()
plt.show()

# Plotting the daily percentage change
plt.figure(figsize=(14, 7))

# GOOGL daily percentage change
plt.subplot(2, 1, 1)
plt.plot(data['Days(GOOGLE)'], data['Daily Change(GOOGLE)'], label='GOOGL', color='b')
plt.title('Daily Percentage Change in Closing Price')
plt.xlabel('Date')
plt.ylabel('Percentage Change (%)')
plt.legend()

# IXIC daily percentage change
plt.subplot(2, 1, 2)
plt.plot(data['Days(IXIC)'], data['Daily Change(IXIC)'], label='IXIC', color='r')
plt.xlabel('Date')
plt.ylabel('Percentage Change (%)')
plt.legend()

plt.tight_layout()
plt.show()
```

The chart shows Google's stock price (GOOGL)  outperforming the Nasdaq Index (IXIC) over the timeframe pictured. While both indexes show a general upward trend, Google’s stock price increased at a steeper rate.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('C:/Users/amith/Downloads/^IXIC.csv')

# Calculate the 30-day moving average for GOOGL
data['30 Day MA(GOOGLE)'] = data['Close(GOOGLE)'].rolling(window=30).mean()

# Plot the closing price and 30-day moving average
plt.figure(figsize=(14, 7))
plt.plot(data['Days(GOOGLE)'], data['Close(GOOGLE)'], label='GOOGL Closing Price', color='b')
plt.plot(data['Days(GOOGLE)'], data['30 Day MA(GOOGLE)'], label='GOOGL 30 Day MA', color='r')
plt.title('GOOGL Stock Price and 30-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Calculate 20-day and 60-day moving averages
data['20-Day MA'] = data['Close(GOOGLE)'].rolling(window=20).mean()
data['60-Day MA'] = data['Close(GOOGLE)'].rolling(window=60).mean()

# Plotting the moving averages
plt.figure(figsize=(14, 7))
plt.plot(data['Days(GOOGLE)'], data['Close(GOOGLE)'], label='GOOGL', color='b')
plt.plot(data['Days(GOOGLE)'], data['20-Day MA'], label='20-Day MA', color='r')
plt.plot(data['Days(GOOGLE)'], data['60-Day MA'], label='60-Day MA', color='g')
plt.title('GOOGL Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
